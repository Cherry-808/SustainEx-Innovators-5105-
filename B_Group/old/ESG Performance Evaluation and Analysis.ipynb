新加坡交易所（SGX）的核心 ESG 指标主要集中在以下几个关键领域，以帮助上市公司更好地披露其环境、社会和治理（ESG）绩效。这些核心指标可被视为制定 ESG 评分系统的基础，涵盖了对投资者和其他利益相关者最具重要性的可持续发展因素：

### 1. **环境（Environmental）**
   - **碳排放**：包括直接和间接碳排放（范围 1、范围 2 和范围 3 排放），以及减排策略。
   - **能源使用**：能源消耗总量，使用可再生能源的比例，以及能源管理措施。
   - **水资源管理**：水资源消耗总量、节水措施及污水处理。
   - **废弃物管理**：废物总量、回收率、有害和无害废弃物的处理。
   - **生物多样性**：对生物多样性和生态系统的影响，如环保区影响及保护措施。

### 2. **社会（Social）**
   - **员工健康与安全**：工伤事故率、健康与安全计划的实施，确保员工在工作环境中的安全。
   - **劳动力多样性与包容性**：员工的性别、年龄、国籍多样性，以及包容性文化的推广措施。
   - **培训与发展**：员工培训和职业发展的投资，提供的培训小时数及效果评估。
   - **社区参与**：公司对当地社区的支持与影响，如志愿者活动、慈善捐赠和社区项目。

### 3. **治理（Governance）**
   - **董事会组成与多样性**：董事会成员的独立性、性别多样性、背景多样性，确保董事会结构能够有效监督公司运营。
   - **道德与合规**：反腐败、反洗钱和反贿赂政策的实施，以及相关的培训与合规管理。
   - **信息披露透明度**：信息披露的频率、及时性和准确性，确保对所有利益相关者的透明度。
   - **股东权利**：确保股东能够参与公司重大决策，股东投票权和股东会议机制。

这些核心指标旨在推动公司有效管理其在可持续发展方面的责任，并为投资者提供一致且有价值的 ESG 信息。

### 1. **定义行业特定 ESG 指标和权重**

在新能源行业，环境模块的权重会较高，尤其是与碳排放、能源效率和资源利用相关的指标。此外，社会和治理方面的权重可以调整，以更好地支持员工安全、技术创新和透明治理。

- **环境（E） - 50%**
  - 碳足迹（25%）：包括范围 1 和范围 2 排放，以及减排措施。新能源行业应特别关注生产过程中的排放以及供应链的碳足迹。
  - 可再生能源比例（15%）：衡量新能源生产过程中自发电和可再生能源的使用比例。
  - 资源效率（5%）：对水资源和材料的管理效率，鼓励循环利用和资源优化。
  - 废弃物管理（5%）：废弃物的处置方式及回收利用情况，尤其关注有害废弃物的处理。

- **社会（S） - 25%**
  - 员工健康与安全（10%）：新能源行业的生产环节可能涉及高温、高压等安全风险，因此需要更严格的安全标准。
  - 劳动力多样性与包容性（5%）：包括性别、年龄和技能的多样性，以支持行业内的创新和包容性文化。
  - 技术研发与员工培训（5%）：鼓励研发投入和技术培训，提升员工技能并促进技术进步。
  - 社区参与（5%）：新能源设施对当地社区的影响，以及对社区可持续发展的支持。

- **治理（G） - 25%**
  - 董事会组成与多样性（10%）：董事会的独立性与多样性，确保在新能源领域的长期战略规划和合规性。
  - 道德与合规（10%）：反腐败和合规管理，特别关注供应链中的道德行为。
  - 信息披露透明度（5%）：如技术创新、环保目标和 ESG 绩效的透明度，支持行业对外界的责任承诺。

---

### 2. **指标评分标准化**

采用针对新能源行业的标准化评分方法：

- **基准评分**：使用行业内的最佳实践和平均值，作为各公司评分的参考基准。
  
- **等级评分**：对可再生能源比例、员工安全措施等指标进行等级划分。例如，自发电和使用可再生能源比例超过 80% 得高分。

- **目标达成度**：依据公司设立的 ESG 目标，达到或超额完成的公司可获得高分。新能源行业通常有排放减少和创新研发目标，达成这些目标可以获得额外评分。

---

### 3. **计算加权总分并生成行业定制的 ESG 报告**

使用加权平均法计算新能源行业的 ESG 总分：

\[
\text{ESG总分} = E \times 50\% + S \times 25\% + G \times 25\%
\]

并为各公司生成详细的行业定制 ESG 报告，其中包括：

- **总体得分与行业基准对比**：清晰展示公司 ESG 总分相对于行业平均值和领先公司的排名。
- **分项分析**：提供环境、社会、治理各模块得分和指标的具体表现，以便于定位改进空间。
- **定制改进建议**：为低分的指标提供行业内最佳实践建议，如如何在能源效率或社区参与方面提升表现。


````python
import pandas as pd

# 定义 ESG 指标及其在新能源行业的权重
weights = {
    "environmental": 0.50,
    "carbon_footprint": 0.25,
    "renewable_energy_ratio": 0.15,
    "resource_efficiency": 0.05,
    "waste_management": 0.05,
    
    "social": 0.25,
    "employee_health_safety": 0.10,
    "diversity_inclusion": 0.05,
    "training_development": 0.05,
    "community_engagement": 0.05,
    
    "governance": 0.25,
    "board_diversity": 0.10,
    "ethics_compliance": 0.10,
    "transparency": 0.05,
}

# 创建一个示例数据集，每行代表一家新能源公司及其各项 ESG 指标得分（0-100）
data = {
    "company": ["Company A", "Company B", "Company C"],
    "carbon_footprint": [80, 60, 70],
    "renewable_energy_ratio": [90, 70, 85],
    "resource_efficiency": [75, 65, 80],
    "waste_management": [70, 85, 75],
    "employee_health_safety": [95, 85, 90],
    "diversity_inclusion": [60, 70, 80],
    "training_development": [85, 90, 88],
    "community_engagement": [70, 65, 80],
    "board_diversity": [80, 75, 85],
    "ethics_compliance": [90, 85, 88],
    "transparency": [80, 75, 82],
}

# 将数据转换为 DataFrame
df = pd.DataFrame(data)

# 定义计算加权得分的函数
def calculate_weighted_score(row, weights):
    # 计算每个模块的得分
    environmental_score = (
        row["carbon_footprint"] * weights["carbon_footprint"] +
        row["renewable_energy_ratio"] * weights["renewable_energy_ratio"] +
        row["resource_efficiency"] * weights["resource_efficiency"] +
        row["waste_management"] * weights["waste_management"]
    ) / weights["environmental"]
    
    social_score = (
        row["employee_health_safety"] * weights["employee_health_safety"] +
        row["diversity_inclusion"] * weights["diversity_inclusion"] +
        row["training_development"] * weights["training_development"] +
        row["community_engagement"] * weights["community_engagement"]
    ) / weights["social"]
    
    governance_score = (
        row["board_diversity"] * weights["board_diversity"] +
        row["ethics_compliance"] * weights["ethics_compliance"] +
        row["transparency"] * weights["transparency"]
    ) / weights["governance"]
    
    # 计算总得分
    total_score = (
        environmental_score * weights["environmental"] +
        social_score * weights["social"] +
        governance_score * weights["governance"]
    )
    
    return total_score, environmental_score, social_score, governance_score

# 计算每家公司 ESG 得分
df[["total_score", "environmental_score", "social_score", "governance_score"]] = df.apply(
    calculate_weighted_score, axis=1, weights=weights, result_type="expand"
)

# 定义一个函数根据总分设置等级
def assign_grade(total_score):
    if total_score >= 70:
        return "Good"
    elif total_score >= 50:
        return "Average"
    else:
        return "Poor"

# 根据总分计算等级
df["grade"] = df["total_score"].apply(assign_grade)

# 输出结果
print(df[["company", "total_score", "environmental_score", "social_score", "governance_score", "grade"]])
````

## Benchmarking and Evaluating ESG Performance

### 1. **数据收集与细分**
   - 收集行业范围内的 ESG 数据，包括多个公司在相同 ESG 指标上的分数。
   - 按照细分行业（如新能源行业的不同子领域）和规模类别（如大中型企业）将数据分组，确保数据具有可比性。

### 2. **计算行业基准**
   - 对每个 ESG 指标，在不同的行业细分和公司规模类别中计算平均值、中位数或百分位数等统计量。例如，可以计算各个细分行业在“碳足迹”上的平均得分作为基准。
   - 如果可能，进一步计算各个子类别中 ESG 总分的分布情况，比如按季度、年度变化趋势。

### 3. **基准与得分标准化**
   - 为使得分具有统一性和可比性，可以对 ESG 得分进行标准化处理，例如通过 Min-Max 标准化或 Z-Score，将公司得分转化为相对于基准的百分位数或差距。
   - 这种处理方式可以消除行业内不同细分和规模对 ESG 分数的影响，使基准更加公平和相关。

### 4. **评估公司相对于基准的绩效**
   - 将个别公司的 ESG 分数与其所在细分行业的基准对比。例如，若一家公司的碳足迹得分高于行业基准，则可视为在该方面表现优于同行。
   - 为每个 ESG 模块分别计算相对绩效得分，例如该公司各模块（环境、社会、治理）相对于行业的相对位置，以更清晰地呈现其综合 ESG 表现。


````python
import pandas as pd

# 示例 ESG 数据，包括公司、细分行业、规模类别及 ESG 各项得分
data = {
    "company": ["Company A", "Company B", "Company C", "Company D"],
    "industry": ["Renewable Energy", "Renewable Energy", "Solar", "Solar"],
    "size": ["Large", "Medium", "Large", "Medium"],
    "carbon_footprint": [80, 60, 70, 85],
    "renewable_energy_ratio": [90, 70, 85, 80],
    "employee_health_safety": [95, 85, 90, 88],
    "board_diversity": [80, 75, 85, 82],
}

# 转换为 DataFrame
df = pd.DataFrame(data)

# 计算每个细分行业和规模类别的 ESG 基准（仅对数值列计算平均值）
numeric_columns = ["carbon_footprint", "renewable_energy_ratio", "employee_health_safety", "board_diversity"]
benchmarks = df.groupby(["industry", "size"])[numeric_columns].mean().reset_index()
benchmarks.rename(columns=lambda x: x + "_benchmark" if x not in ["industry", "size"] else x, inplace=True)

# 将公司数据与基准数据合并
df = pd.merge(df, benchmarks, on=["industry", "size"], how="left")

# 计算相对绩效
for column in numeric_columns:
    df[f"{column}_relative_performance"] = df[column] - df[f"{column}_benchmark"]

# 输出结果
print(df[["company", "industry", "size", "carbon_footprint_relative_performance",
          "renewable_energy_ratio_relative_performance", "employee_health_safety_relative_performance",
          "board_diversity_relative_performance"]])
````

为了探索 **ESG 绩效** 与 **财务指标**（如 **股本回报率 (ROE)** 和 **股票表现**）之间的联系，可以使用 **相关性分析** 或 **回归分析** 方法。通过这些分析，我们可以揭示 ESG 分数如何与公司盈利能力或风险相关，为投资者和管理人员提供有关 ESG 实践对财务影响的见解。

### 1. **数据准备**
假设数据集中包含：
- **公司名称**、**年份**
- **ESG 分数**
- **股本回报率 (ROE)**、**股票表现 (如年收益率)** 等财务指标

### 2. **相关性分析**
使用 **皮尔逊相关系数**（Pearson correlation coefficient）来衡量 ESG 分数与财务指标之间的线性相关性。通过计算这些指标的相关性，可以了解 ESG 绩效与财务表现之间的关系强度。

### 3. **回归分析**
可以使用 **线性回归** 或 **多元回归分析** 来评估 ESG 分数是否对财务表现（如 ROE 或股票表现）有显著的预测作用。

### 4. **实现步骤**

#### 示例数据假设：
- **esg_score**：ESG 综合得分
- **roe**：股本回报率（Return on Equity）
- **stock_performance**：股票年收益率（Stock return）

### 6. **代码解释**

1. **相关性分析**
   - 使用 `df.corr()` 计算数据框中各列之间的皮尔逊相关系数，特别是 **ESG 分数** 与 **股本回报率 (ROE)** 和 **股票表现** 之间的相关性。
   - `sns.heatmap()` 用于绘制相关性矩阵热图，便于直观地查看各变量之间的相关关系。

2. **回归分析**
   - 使用 `statsmodels.api.OLS` 进行 **普通最小二乘法 (OLS)** 回归分析，评估 **ESG 分数** 对 **ROE** 和 **股票表现** 的影响。
   - `model.summary()` 输出回归模型的统计结果，包括 R²、p 值等关键信息，帮助判断 ESG 分数与财务指标的关系是否显著。

3. **可视化**
   - 使用 `sns.regplot()` 绘制回归线，显示 **ESG 分数** 与 **ROE** 和 **股票表现** 之间的关系。

### 7. **输出分析**
1. **相关性分析**
   - 输出的相关系数表明 ESG 分数与财务指标（ROE 和股票表现）之间的线性关系强度。
   - 例如，若 ESG 分数与 ROE 之间的相关系数接近 1，表明两者有强正相关。

2. **回归分析**
   - 回归结果会提供一个 R² 值，表示回归模型能够解释财务指标波动的程度。
   - 如果回归模型的 p 值小于显著性水平（通常为 0.05），则 ESG 分数对财务指标的影响是显著的。

### 8. **可能的结论**
- **高 ESG 得分与财务表现的关系**：如果回归分析结果表明 ESG 得分与 ROE 或股票表现之间存在显著的正相关关系，那么企业的 ESG 实践可能对其盈利能力或投资回报产生积极影响。
- **负相关或无显著关系**：如果相关性较低或回归分析显示 ESG 得分对财务指标没有显著预测能力，那么 ESG 绩效可能不直接影响公司的财务表现。

### 总结
通过相关性分析和回归分析，我们能够评估 ESG 绩效与财务指标之间的关系，为投资者和公司管理层提供关于 ESG 实践财务影响的重要见解。这些分析有助于更好地理解 ESG 投资决策的潜在财务收益或风险。

````python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import statsmodels.api as sm

# 假设数据集
data = {
    "company": ["Company A", "Company B", "Company C", "Company D"],
    "esg_score": [80, 60, 75, 90],
    "roe": [15, 10, 12, 18],
    "stock_performance": [10, 5, 7, 12],
}

# 转换为 DataFrame
df = pd.DataFrame(data)

# 1. 计算 ESG 分数与财务指标的相关性（过滤出数值列）
numeric_df = df.select_dtypes(include=["number"])  # 只选择数值列
correlation_matrix = numeric_df.corr()
print("Correlation Matrix:\n", correlation_matrix)

# 可视化相关性矩阵
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Matrix between ESG Score and Financial Indicators")
plt.show()

# 2. 回归分析：ESG Score 对 ROE 的影响
X = df["esg_score"]  # ESG 作为自变量
y = df["roe"]  # ROE 作为因变量
X = sm.add_constant(X)  # 加入常数项

model = sm.OLS(y, X).fit()  # OLS回归模型
print(model.summary())  # 输出回归结果

# 3. 回归分析：ESG Score 对股票年收益率的影响
y_stock = df["stock_performance"]
model_stock = sm.OLS(y_stock, X).fit()  # OLS回归模型
print(model_stock.summary())  # 输出回归结果

# 可视化回归结果
sns.regplot(x="esg_score", y="roe", data=df, line_kws={"color": "red"})
plt.title("Regression: ESG Score vs ROE")
plt.show()

sns.regplot(x="esg_score", y="stock_performance", data=df, line_kws={"color": "red"})
plt.title("Regression: ESG Score vs Stock Performance")
plt.show()
````
